```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
print(df_market_data.shape)

# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_data =StandardScaler().fit_transform(df_market_data)
scaled_data
```

```python
df_market_data.columns
```

```python
# Create a DataFrame with the scaled data
scaled_df =pd.DataFrame(scaled_data, columns=df_market_data.columns)
scaled_df
```

```python
#  Copy the crypto names from the original data
crypto_names = df_market_data.index
crypto_names
```

```python
scaled_df['coin_id']=crypto_names
# Set the coinid column as index make new column
scaled_df.set_index('coin_id', drop=True, inplace=True)
# Display sample data
scaled_df
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values= list(range(1,12))
k_values
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for k in k_values:
    print(k)
    # 1. Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=k, random_state=1)
    # 2. Fit the model to the data using `df_market_data_scaled`
    k_model.fit(scaled_df)
    # 3. Append the model.inertia_ to the inertia list
    inertia_values.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia":inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df =pd.DataFrame(elbow_data)
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df.hvplot.line(
    x="k",
    y= "inertia",
    title= "Elbow Curve 1",
    xticks =k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** the best value of k=4 which is 79.022 

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model =  KMeans(n_clusters = 4,random_state = 1)
model
```

```python
# Fit the K-Means model using the scaled data
model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4= model.predict(scaled_df)

# Print the resulting array of cluster values.
k_4
```

```python
# Create a copy of the DataFrame
copy_scaled_df= scaled_df.copy()
copy_scaled_df
```

```python
# Add a new column to the DataFrame with the predicted clusters
copy_scaled_df['predicted_clusters']= k_4
# Display sample data
copy_scaled_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# Plot the clusters
copy_scaled_df.hvplot.scatter(
    # `x="price_change_percentage_24h"
    x="price_change_percentage_24h",
    # y="price_change_percentage_7d"`.
    y="price_change_percentage_7d",
    by="predicted_clusters",
    hover_cols ="coin_id",
    title= "Predicted clusters"
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca= PCA(n_components=3)
pca
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
clusters_pca = pca.fit_transform(copy_scaled_df)
# View the first five rows of the DataFrame. 
clusters_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```
